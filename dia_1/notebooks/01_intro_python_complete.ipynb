## 4. Listas, Tuplas, Sets y Diccionarios

| Tipo | Mutable | Ordenado | Duplicados | Sintaxis |
|------|---------|----------|------------|----------|
| `list`  | Sí | Sí | Sí | `[a, b]` |
| `tuple` | No | Sí | Sí | `(a, b)` |
| `set`   | Sí | No | No | `{a, b}` |
| `dict`  | Sí | Sí (3.7+) | Claves no | `{k: v}` |

````python
frutas = ['manzana', 'pera', 'naranja', 'kiwi', 'mango']

print(frutas[0], frutas[-1])
print(frutas[1:3])

frutas.append('fresa')
frutas.insert(0, 'uva')
frutas.remove('kiwi')
ultima = frutas.pop()
print(frutas)
print(sorted(frutas))
print('mango' in frutas)
````

````python
# List comprehensions
cuadrados = [i**2 for i in range(10)]
print(cuadrados)

pares = [i for i in range(20) if i % 2 == 0]
print(pares)

mayusculas = [f.upper() for f in frutas]
print(mayusculas)
````

````python
# Tuplas: inmutables
rgb_rojo = (255, 0, 0)
r, g, b = rgb_rojo          # desempaquetado
print(f'R={r}, G={g}, B={b}')

try:
    rgb_rojo[0] = 128
except TypeError as e:
    print(f'Error esperado: {e}')
````

````python
# Sets: sin duplicados, operaciones de conjuntos
python_devs = {'Ana', 'Luis', 'Marta', 'Juan', 'Ana'}  # Ana duplicada
gis_devs    = {'Luis', 'Marta', 'Sara', 'Pedro'}

print(python_devs)             # sin duplicados
print(python_devs & gis_devs)  # intersección
print(python_devs | gis_devs)  # unión
print(python_devs - gis_devs)  # diferencia
````

````python
# Diccionarios
persona = {
    'nombre': 'Ana García',
    'edad': 32,
    'lenguajes': ['Python', 'R', 'SQL']
}

print(persona['nombre'])
print(persona.get('email', 'No disponible'))  # .get() con valor por defecto

persona['email'] = 'ana@example.com'          # añadir clave

for clave, valor in persona.items():
    print(f'  {clave}: {valor}')
````

````python
# Dict comprehension + zip + enumerate
notas = {'Ana': 9.2, 'Luis': 7.8, 'Marta': 8.5, 'Juan': 6.3}

aprobados = {n: v for n, v in notas.items() if v >= 7}
print(aprobados)

nombres  = ['Ana', 'Luis', 'Marta']
ciudades = ['Madrid', 'Sevilla', 'Barcelona']

for i, (nombre, ciudad) in enumerate(zip(nombres, ciudades)):
    print(f'[{i}] {nombre} → {ciudad}')
````

## 6. Condicionales

````python
nota = 7.3

if nota >= 9:
    calificacion = 'Sobresaliente'
elif nota >= 7:
    calificacion = 'Notable'
elif nota >= 5:
    calificacion = 'Aprobado'
else:
    calificacion = 'Suspenso'

print(f'{nota} → {calificacion}')
````

````python
# match/case (Python 3.10+)
def describir_tipo(obj):
    match type(obj).__name__:
        case 'int':   return f'{obj} es un entero'
        case 'float': return f'{obj} es un flotante'
        case 'str':   return f'"{obj}" es un string de {len(obj)} caracteres'
        case 'list':  return f'Una lista con {len(obj)} elementos'
        case _:       return f'Tipo: {type(obj).__name__}'

for obj in [42, 3.14, 'hola', [1, 2, 3], {'a': 1}]:
    print(describir_tipo(obj))
````

## 7. Funciones

Las funciones son bloques de código reutilizable. Son la base para evitar repeticiones y estructurar programas.

````python
def saludar(nombre: str, veces: int = 1) -> str:
    """
    Genera un saludo personalizado.

    Args:
        nombre: nombre de la persona
        veces: número de repeticiones

    Returns:
        str: el saludo
    """
    return ' '.join([f'¡Hola, {nombre}!'] * veces)


print(saludar('Ana'))
print(saludar('Luis', veces=3))
help(saludar)
````

````python
# Múltiples valores de retorno (en realidad una tupla)
def minmax(lista: list) -> tuple:
    return min(lista), max(lista)


valores = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3]
minimo, maximo = minmax(valores)
print(f'Min: {minimo}, Max: {maximo}')
````

````python
# *args: número variable de argumentos
def resumir(*args, titulo='Resumen', separador='-'):
    print(titulo)
    print(separador * len(titulo))
    for i, elem in enumerate(args, 1):
        print(f'  {i}. {elem}')


resumir('Python', 'GeoPandas', 'Rasterio', 'OSMnx', titulo='Temario GeoPython 2026')
````

````python
# Variables locales vs globales
x = 'global'

def modificar_local():
    x = 'local'          # no afecta a la global
    print(f'Dentro: {x}')

modificar_local()
print(f'Fuera: {x}')     # sigue siendo 'global'
````

````python
# Lambdas: funciones anónimas de una línea
palabras = ['banana', 'kiwi', 'manzana', 'uva', 'cereza']

print(sorted(palabras, key=lambda p: len(p)))                     # por longitud
print(list(filter(lambda p: len(p) > 5, palabras)))               # más de 5 letras
print(list(map(lambda p: p.upper(), palabras)))                   # todas en mayúsculas
````

## 8. Manejo de errores

`try/except` evita que el programa se detenga al encontrar un error.

````python
datos = [10, 0, 5, 'texto', 8, None, 2]
resultados = []

for d in datos:
    try:
        resultados.append(100 / d)
    except ZeroDivisionError:
        print(f'División por cero')
    except TypeError as e:
        print(f'Tipo incorrecto ({d!r}): {e}')

print(resultados)
````

````python
# try / except / else / finally
def dividir(a, b):
    try:
        resultado = a / b
    except ZeroDivisionError:
        print('Error: división por cero')
        return None
    else:
        return resultado       # solo si NO hubo excepción
    finally:
        print('Esto se ejecuta siempre')


print(dividir(10, 2))
print()
print(dividir(10, 0))
````

````python
# Lanzar nuestras propias excepciones
def calcular_media(lista):
    if not lista:
        raise ValueError('La lista no puede estar vacía')
    if not all(isinstance(x, (int, float)) for x in lista):
        raise TypeError('Todos los elementos deben ser numéricos')
    return sum(lista) / len(lista)


for caso in [[1, 2, 3], [], [1, 2, 'tres']]:
    try:
        print(f'Media de {caso}: {calcular_media(caso)}')
    except (ValueError, TypeError) as e:
        print(f'Error: {e}')
````

## 9. Módulos y archivos

````python
%%writefile utilidades.py
"""
Módulo de utilidades — GeoPython 2026.
"""

def normalizar(valores):
    """Normaliza una lista al rango [0, 1]."""
    vmin, vmax = min(valores), max(valores)
    if vmin == vmax:
        return [0.0] * len(valores)
    return [(v - vmin) / (vmax - vmin) for v in valores]


def estadisticas(valores):
    """Devuelve media, min, max y desviación estándar."""
    import math
    n = len(valores)
    media = sum(valores) / n
    varianza = sum((v - media)**2 for v in valores) / n
    return {'n': n, 'media': round(media, 4),
            'min': min(valores), 'max': max(valores),
            'std': round(math.sqrt(varianza), 4)}
````

````python
import utilidades as ut

datos = [3, 7, 2, 9, 1, 5, 8, 4, 6]
print(ut.normalizar(datos))
print(ut.estadisticas(datos))
````
