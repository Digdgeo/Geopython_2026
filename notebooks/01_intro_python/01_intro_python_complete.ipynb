```python
# Tuplas: inmutables, útiles para coordenadas y valores fijos
punto = (37.3891, -5.9845)  # coordenadas Sevilla
print(f'Latitud: {punto[0]}, Longitud: {punto[1]}')

# Desempaquetado
lat, lon = punto
print(f'lat={lat}, lon={lon}')

# Las tuplas son inmutables: esto genera error
try:
    punto[0] = 0
except TypeError as e:
    print(f'Error esperado: {e}')
```

```python
# Sets: colecciones sin duplicados, útiles para operaciones de conjuntos
sensores_a = {'S2', 'L8', 'L9', 'S1'}
sensores_b = {'S2', 'L9', 'MODIS'}

print(f'Intersección: {sensores_a & sensores_b}')
print(f'Unión: {sensores_a | sensores_b}')
print(f'Diferencia A-B: {sensores_a - sensores_b}')
print(f'Diferencia simétrica: {sensores_a ^ sensores_b}')
```

```python
# Diccionarios: pares clave-valor
isla_info = {
    'nombre': 'Tenerife',
    'area_km2': 2034,
    'poblacion': 928604,
    'capital': 'Santa Cruz de Tenerife',
    'epsg': 32628
}

print(isla_info['nombre'])
print(isla_info.get('altura_max', 'No disponible'))  # .get() con valor por defecto

# Añadir y modificar
isla_info['altura_max'] = 3715  # Teide
print(isla_info)
```

```python
# match/case (Python 3.10+): alternativa más limpia al if/elif largo
def clasificar_sensor(sensor):
    match sensor:
        case 'S2':
            return 'Sentinel-2 (ESA) — 10m resolución'
        case 'L8' | 'L9':
            return 'Landsat 8/9 (USGS) — 30m resolución'
        case 'MODIS':
            return 'MODIS (NASA) — 250m-1km resolución'
        case _:
            return f'Sensor desconocido: {sensor}'

for s in ['S2', 'L8', 'L9', 'MODIS', 'SPOT']:
    print(clasificar_sensor(s))
```

## 7. Funciones

Las funciones son bloques de código reutilizable. En Python se definen con `def`.

```python
# Función básica con docstring
def calcular_ndvi(nir, red):
    """
    Calcula el NDVI (Normalized Difference Vegetation Index).
    
    Args:
        nir (float): Reflectancia en el infrarrojo cercano (banda NIR)
        red (float): Reflectancia en el rojo (banda RED)
    
    Returns:
        float: Valor NDVI en el rango [-1, 1]
    
    Raises:
        ValueError: Si nir + red == 0 (evita división por cero)
    """
    if (nir + red) == 0:
        raise ValueError('nir + red no puede ser 0')
    return (nir - red) / (nir + red)


print(calcular_ndvi(0.4, 0.1))
print(calcular_ndvi(0.05, 0.3))  # agua/suelo
```

```python
# La docstring es accesible con help() o ?
help(calcular_ndvi)
```

```python
# Parámetros por defecto y keyword arguments
def calcular_indice(banda1, banda2, nombre='indice', decimales=4):
    """
    Calcula un índice de diferencia normalizada genérico.
    """
    resultado = (banda1 - banda2) / (banda1 + banda2)
    return round(resultado, decimales)


# Distintas formas de llamarla
print(calcular_indice(0.4, 0.1))                           # args posicionales
print(calcular_indice(0.4, 0.1, nombre='NDVI'))            # con keyword
print(calcular_indice(banda2=0.1, banda1=0.4, decimales=2))# orden inverso
```

```python
# Funciones que devuelven múltiples valores (en realidad una tupla)
def estadisticas_basicas(valores):
    """
    Devuelve media, mínimo y máximo de una lista de valores.
    """
    media = sum(valores) / len(valores)
    return media, min(valores), max(valores)


ndvis = [0.12, 0.45, 0.67, 0.33, 0.78, 0.55]
media, minimo, maximo = estadisticas_basicas(ndvis)
print(f'Media: {media:.3f}, Min: {minimo}, Max: {maximo}')
```

```python
# Funciones con número variable de argumentos
def combinar_bandas(*args, separador='_'):
    """Combina nombres de bandas en un string."""
    return separador.join(args)


print(combinar_bandas('B4', 'B8'))           # 2 bandas
print(combinar_bandas('B2', 'B3', 'B4'))     # 3 bandas
print(combinar_bandas('B4', 'B8', separador='-'))  # con separador
```

```python
# Funciones lambda: funciones anónimas de una línea
# Útiles para operaciones rápidas, map, filter, sorted

ndvi_clase = lambda v: 'densa' if v > 0.5 else ('moderada' if v > 0.2 else 'escasa')

valores = [0.1, 0.35, 0.72]
print(list(map(ndvi_clase, valores)))

# Ordenar una lista de tuplas por el segundo elemento
puntos = [('A', 0.65), ('B', 0.12), ('C', 0.88)]
ordenados = sorted(puntos, key=lambda p: p[1], reverse=True)
print(ordenados)
```

## 9. Manejo de errores

Los errores en Python son objetos de tipo `Exception`. Podemos capturarlos con `try/except` para que el programa no se interrumpa inesperadamente.

```python
# Sin manejo de errores: el programa se rompe
valores = [1, 2, 3, 'nodata', 5, 6]
resultado = []

for v in valores:
    try:
        resultado.append(v * 2)
    except TypeError:
        print(f'Valor no numérico ignorado: {v!r}')

print(resultado)
```

```python
# Capturar múltiples tipos de error
def leer_valor(diccionario, clave):
    try:
        valor = diccionario[clave]
        return float(valor)
    except KeyError:
        print(f'Clave no encontrada: {clave!r}')
        return None
    except (ValueError, TypeError) as e:
        print(f'No se puede convertir a float: {e}')
        return None
    finally:
        pass  # este bloque se ejecuta siempre (útil para cerrar archivos)


datos = {'ndvi': '0.65', 'clase': 'densa', 'area': None}
print(leer_valor(datos, 'ndvi'))
print(leer_valor(datos, 'clase'))
print(leer_valor(datos, 'area'))
print(leer_valor(datos, 'missing'))
```

```python
# Lanzar nuestras propias excepciones
def validar_ndvi(valor):
    """Valida que el NDVI esté en el rango correcto."""
    if not isinstance(valor, (int, float)):
        raise TypeError(f'NDVI debe ser numérico, recibido: {type(valor).__name__}')
    if not -1 <= valor <= 1:
        raise ValueError(f'NDVI fuera de rango [-1, 1]: {valor}')
    return valor


for v in [0.5, 1.5, 'texto']:
    try:
        print(f'NDVI válido: {validar_ndvi(v)}')
    except (ValueError, TypeError) as e:
        print(f'Error: {e}')
```

## 10. Módulos

Un módulo es un archivo `.py` que contiene funciones, clases o variables que podemos importar en otro script o notebook.

```python
# Creamos un módulo con funciones útiles para el curso
%%writefile indices_espectrales.py
"""
Módulo con funciones para calcular índices espectrales.
GeoPython 2026
"""

def ndvi(nir: float, red: float) -> float:
    """Normalized Difference Vegetation Index."""
    if (nir + red) == 0:
        return 0.0
    return (nir - red) / (nir + red)


def ndwi(green: float, nir: float) -> float:
    """Normalized Difference Water Index (Gao 1996)."""
    if (green + nir) == 0:
        return 0.0
    return (green - nir) / (green + nir)


def nbr(nir: float, swir: float) -> float:
    """Normalized Burn Ratio."""
    if (nir + swir) == 0:
        return 0.0
    return (nir - swir) / (nir + swir)


INDICES = {
    'ndvi': ndvi,
    'ndwi': ndwi,
    'nbr': nbr,
}
```

```python
# Importar el módulo
import indices_espectrales as ie

print(ie.ndvi(0.4, 0.1))
print(ie.ndwi(0.15, 0.4))
print(ie.nbr(0.4, 0.2))

# Usar el diccionario de funciones
for nombre, funcion in ie.INDICES.items():
    print(f'{nombre.upper()}: {funcion(0.4, 0.1):.4f}')
```

## 12. Clases

Las clases son la base de la **Programación Orientada a Objetos (POO)**.  
Permiten encapsular datos y comportamiento en una misma unidad.

Esto es relevante para el curso porque librerías como GeoPandas o Rasterio están construidas sobre clases: cuando abrimos un shapefile con `gpd.read_file()` obtenemos un objeto `GeoDataFrame` con sus propios métodos y atributos.

```python
# Clase básica: Punto geográfico
class PuntoGeo:
    """
    Representa un punto geográfico con latitud, longitud y nombre opcional.
    """

    def __init__(self, lat: float, lon: float, nombre: str = 'Sin nombre'):
        """
        Constructor: se ejecuta al crear una instancia.
        
        Args:
            lat: latitud en grados decimales
            lon: longitud en grados decimales
            nombre: nombre descriptivo del punto
        """
        self.lat = lat
        self.lon = lon
        self.nombre = nombre

    def __repr__(self):
        """Representación técnica del objeto."""
        return f'PuntoGeo({self.lat}, {self.lon}, "{self.nombre}")'

    def __str__(self):
        """Representación legible del objeto."""
        return f'{self.nombre} ({self.lat:.4f}N, {self.lon:.4f}E)'

    def wkt(self) -> str:
        """Devuelve el punto en formato WKT."""
        return f'POINT({self.lon} {self.lat})'

    def distancia(self, otro: 'PuntoGeo') -> float:
        """
        Distancia euclídea aproximada (solo para distancias cortas).
        Para distancias reales usar pyproj o geopy.
        """
        return math.sqrt((self.lat - otro.lat)**2 + (self.lon - otro.lon)**2)


# Crear instancias
ebd = PuntoGeo(37.3891, -5.9845, 'Estación Biológica de Doñana')
teide = PuntoGeo(28.2723, -16.6431, 'Teide')

print(ebd)              # usa __str__
print(repr(ebd))        # usa __repr__
print(ebd.wkt())
print(f'Distancia aprox: {ebd.distancia(teide):.4f}°')
```

```python
# Herencia: crear una clase que extiende PuntoGeo
class EstacionMeteo(PuntoGeo):
    """
    Estación meteorológica: extiende PuntoGeo añadiendo datos de temperatura.
    """

    def __init__(self, lat: float, lon: float, nombre: str, red: str = 'AEMET'):
        super().__init__(lat, lon, nombre)   # llamar al constructor del padre
        self.red = red
        self.registros = []                  # lista de (fecha, temp)

    def __str__(self):
        return f'[{self.red}] {super().__str__()}'

    def registrar_temperatura(self, fecha: str, temp: float):
        """Añade un registro de temperatura."""
        self.registros.append((fecha, temp))

    def temperatura_media(self) -> float | None:
        """Calcula la temperatura media de todos los registros."""
        if not self.registros:
            return None
        return sum(t for _, t in self.registros) / len(self.registros)

    def resumen(self):
        """Imprime un resumen de la estación."""
        print(f'{self}')
        print(f'  WKT: {self.wkt()}')
        print(f'  Registros: {len(self.registros)}')
        if self.registros:
            print(f'  Temperatura media: {self.temperatura_media():.1f}°C')


# Crear y usar una estación
estacion = EstacionMeteo(37.42, -5.90, 'Sevilla Airport', red='AEMET')
for fecha, temp in [('2026-04-01', 22.3), ('2026-04-02', 24.1), ('2026-04-03', 19.8)]:
    estacion.registrar_temperatura(fecha, temp)

estacion.resumen()

# isinstance verifica si un objeto pertenece a una clase (o a su jerarquía)
print(f'\n¿Es EstacionMeteo? {isinstance(estacion, EstacionMeteo)}')
print(f'¿Es PuntoGeo?       {isinstance(estacion, PuntoGeo)}')  # True por herencia
```

```python
# Atributos y métodos de clase (compartidos por todas las instancias)
class ImagenSatelite:
    """
    Clase base para imágenes de satélite. Veremos una versión más completa
    en el módulo 5 cuando trabajemos con Rasterio.
    """
    
    SENSORES_SOPORTADOS = ['Landsat-8', 'Landsat-9', 'Sentinel-2']  # atributo de clase
    _contador = 0                                                     # privado por convención

    def __init__(self, sensor: str, fecha: str, path_archivo: str):
        if sensor not in self.SENSORES_SOPORTADOS:
            raise ValueError(f'Sensor no soportado: {sensor}. Usar: {self.SENSORES_SOPORTADOS}')
        self.sensor = sensor
        self.fecha = fecha
        self.path = path_archivo
        ImagenSatelite._contador += 1
        self.id = ImagenSatelite._contador

    def __str__(self):
        return f'[{self.id}] {self.sensor} — {self.fecha}'

    @classmethod
    def total_imagenes(cls) -> int:
        """Método de clase: accede a cls en vez de self."""
        return cls._contador

    @staticmethod
    def es_fecha_valida(fecha: str) -> bool:
        """Método estático: no accede ni a self ni a cls."""
        partes = fecha.split('-')
        return len(partes) == 3 and all(p.isdigit() for p in partes)


img1 = ImagenSatelite('Landsat-8', '2026-03-15', '/data/L8_20260315.tif')
img2 = ImagenSatelite('Sentinel-2', '2026-03-20', '/data/S2_20260320.tif')

print(img1)
print(img2)
print(f'Total imágenes creadas: {ImagenSatelite.total_imagenes()}')
print(f'Fecha válida "2026-03-15": {ImagenSatelite.es_fecha_valida("2026-03-15")}')
print(f'Fecha válida "marzo/2026": {ImagenSatelite.es_fecha_valida("marzo/2026")}')
```

## 13. Ejercicio final: Rock, Paper, Scissors, Lizard, Spock

El juego popularizado por The Big Bang Theory. Las reglas:

- Las tijeras cortan el papel
- El papel envuelve a la piedra  
- La piedra aplasta al lagarto
- El lagarto envenena a Spock
- Spock destruye las tijeras
- Las tijeras decapitan al lagarto
- El lagarto come el papel
- El papel refuta a Spock
- Spock vaporiza la piedra
- La piedra aplasta las tijeras

Lo usamos como excusa para practicar todo lo visto: funciones, diccionarios, condicionales, clases y bucles.

---

```python
# Versión funcional con diccionario (sin if/elif largo)
import random

OPCIONES = ['piedra', 'papel', 'tijeras', 'lagarto', 'spock']

# Diccionario: clave = (ganador, perdedor) → descripción de la victoria
REGLAS = {
    ('tijeras', 'papel'):   'Las tijeras cortan el papel',
    ('papel',   'piedra'):  'El papel envuelve la piedra',
    ('piedra',  'lagarto'): 'La piedra aplasta al lagarto',
    ('lagarto', 'spock'):   'El lagarto envenena a Spock',
    ('spock',   'tijeras'): 'Spock destruye las tijeras',
    ('tijeras', 'lagarto'): 'Las tijeras decapitan al lagarto',
    ('lagarto', 'papel'):   'El lagarto come el papel',
    ('papel',   'spock'):   'El papel refuta a Spock',
    ('spock',   'piedra'):  'Spock vaporiza la piedra',
    ('piedra',  'tijeras'): 'La piedra aplasta las tijeras',
}


def jugada_cpu() -> str:
    """Elige una opción aleatoria para la CPU."""
    return random.choice(OPCIONES)


def determinar_ganador(jugador: str, cpu: str) -> str:
    """
    Determina el resultado de una ronda.
    
    Returns:
        'jugador', 'cpu' o 'empate'
    """
    if jugador == cpu:
        return 'empate'
    elif (jugador, cpu) in REGLAS:
        return 'jugador'
    else:
        return 'cpu'


def jugar_ronda(eleccion_jugador: str) -> dict:
    """
    Ejecuta una ronda completa.
    
    Args:
        eleccion_jugador: elección del jugador (debe estar en OPCIONES)
    
    Returns:
        dict con las claves: jugador, cpu, resultado, descripcion
    """
    eleccion_jugador = eleccion_jugador.lower().strip()
    if eleccion_jugador not in OPCIONES:
        raise ValueError(f'Opción no válida. Elige entre: {OPCIONES}')

    eleccion_cpu = jugada_cpu()
    resultado = determinar_ganador(eleccion_jugador, eleccion_cpu)

    if resultado == 'empate':
        descripcion = '¡Empate!'
    elif resultado == 'jugador':
        descripcion = REGLAS[(eleccion_jugador, eleccion_cpu)] + ' → ¡Ganas!'
    else:
        descripcion = REGLAS[(eleccion_cpu, eleccion_jugador)] + ' → Pierde la CPU... espera, tú pierdes.'

    return {
        'jugador': eleccion_jugador,
        'cpu': eleccion_cpu,
        'resultado': resultado,
        'descripcion': descripcion
    }


# Jugar varias rondas de prueba
for eleccion in ['piedra', 'spock', 'lagarto', 'tijeras', 'papel']:
    ronda = jugar_ronda(eleccion)
    print(f'Tú: {ronda["jugador"]:8s} | CPU: {ronda["cpu"]:8s} | {ronda["descripcion"]}')
```

```python
# Versión orientada a objetos: clase Partida
class Partida:
    """
    Gestiona una partida de Rock-Paper-Scissors-Lizard-Spock.
    Lleva el marcador y el historial de rondas.
    """

    def __init__(self, nombre_jugador: str = 'Jugador'):
        self.nombre = nombre_jugador
        self.victorias = 0
        self.derrotas = 0
        self.empates = 0
        self.historial = []

    @property
    def total_rondas(self) -> int:
        return self.victorias + self.derrotas + self.empates

    @property
    def porcentaje_victorias(self) -> float:
        if self.total_rondas == 0:
            return 0.0
        return self.victorias / self.total_rondas * 100

    def jugar(self, eleccion: str) -> None:
        """Juega una ronda y actualiza el marcador."""
        ronda = jugar_ronda(eleccion)
        self.historial.append(ronda)

        if ronda['resultado'] == 'jugador':
            self.victorias += 1
        elif ronda['resultado'] == 'cpu':
            self.derrotas += 1
        else:
            self.empates += 1

        print(f'Tú: {ronda["jugador"]:8s} | CPU: {ronda["cpu"]:8s} | {ronda["descripcion"]}')

    def marcador(self) -> None:
        """Muestra el marcador actual."""
        print(f'\n=== Marcador {self.nombre} ===')
        print(f'  Victorias: {self.victorias}')
        print(f'  Derrotas:  {self.derrotas}')
        print(f'  Empates:   {self.empates}')
        print(f'  Total:     {self.total_rondas}')
        print(f'  Win rate:  {self.porcentaje_victorias:.1f}%')


# Jugar una partida automática
random.seed(42)  # para reproducibilidad
partida = Partida('Sheldon')

for eleccion in ['spock', 'piedra', 'lagarto', 'tijeras', 'papel', 'spock', 'spock', 'lagarto']:
    partida.jugar(eleccion)

partida.marcador()
```
