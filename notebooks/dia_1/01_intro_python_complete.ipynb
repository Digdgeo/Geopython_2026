| Tipo | Mutable | Ordenado | Duplicados | Sintaxis |

````python
frutas = ['manzana', 'pera', 'naranja', 'kiwi', 'mango']

print(frutas[0], frutas[-1])
print(frutas[1:3])

frutas.append('fresa')
frutas.insert(0, 'uva')
frutas.remove('kiwi')
ultima = frutas.pop()
print(frutas)
print(sorted(frutas))
print('mango' in frutas)
````

````python
# List comprehensions
cuadrados = [i**2 for i in range(10)]
print(cuadrados)

pares = [i for i in range(20) if i % 2 == 0]
print(pares)

mayusculas = [f.upper() for f in frutas]
print(mayusculas)
````

````python
# Tuplas: inmutables
rgb_rojo = (255, 0, 0)
r, g, b = rgb_rojo          # desempaquetado
print(f'R={r}, G={g}, B={b}')

try:
    rgb_rojo[0] = 128
except TypeError as e:
    print(f'Error esperado: {e}')
````

````python
# Sets: sin duplicados, operaciones de conjuntos
python_devs = {'Ana', 'Luis', 'Marta', 'Juan', 'Ana'}  # Ana duplicada
gis_devs    = {'Luis', 'Marta', 'Sara', 'Pedro'}

print(python_devs)             # sin duplicados
print(python_devs & gis_devs)  # intersección
print(python_devs | gis_devs)  # unión
print(python_devs - gis_devs)  # diferencia
````

````python
# Diccionarios
persona = {
    'nombre': 'Ana García',
    'edad': 32,
    'lenguajes': ['Python', 'R', 'SQL']
}

print(persona['nombre'])
print(persona.get('email', 'No disponible'))  # .get() con valor por defecto

persona['email'] = 'ana@example.com'          # añadir clave

for clave, valor in persona.items():
    print(f'  {clave}: {valor}')
````

````python
# Dict comprehension + zip + enumerate
notas = {'Ana': 9.2, 'Luis': 7.8, 'Marta': 8.5, 'Juan': 6.3}

aprobados = {n: v for n, v in notas.items() if v >= 7}
print(aprobados)

nombres  = ['Ana', 'Luis', 'Marta']
ciudades = ['Madrid', 'Sevilla', 'Barcelona']

for i, (nombre, ciudad) in enumerate(zip(nombres, ciudades)):
    print(f'[{i}] {nombre} → {ciudad}')
````

````python
# match/case (Python 3.10+)
def describir_tipo(obj):
    match type(obj).__name__:
        case 'int':   return f'{obj} es un entero'
        case 'float': return f'{obj} es un flotante'
        case 'str':   return f'"{obj}" es un string de {len(obj)} caracteres'
        case 'list':  return f'Una lista con {len(obj)} elementos'
        case _:       return f'Tipo: {type(obj).__name__}'

for obj in [42, 3.14, 'hola', [1, 2, 3], {'a': 1}]:
    print(describir_tipo(obj))
````

## 7. Funciones

Las funciones son bloques de código reutilizable. Son la base para evitar repeticiones y estructurar programas.

````python
def saludar(nombre: str, veces: int = 1) -> str:
    """
    Genera un saludo personalizado.

    Args:
        nombre: nombre de la persona
        veces: número de repeticiones

    Returns:
        str: el saludo
    """
    return ' '.join([f'¡Hola, {nombre}!'] * veces)


print(saludar('Ana'))
print(saludar('Luis', veces=3))
help(saludar)
````

````python
# Múltiples valores de retorno (en realidad una tupla)
def minmax(lista: list) -> tuple:
    return min(lista), max(lista)


valores = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3]
minimo, maximo = minmax(valores)
print(f'Min: {minimo}, Max: {maximo}')
````

````python
# *args: número variable de argumentos
def resumir(*args, titulo='Resumen', separador='-'):
    print(titulo)
    print(separador * len(titulo))
    for i, elem in enumerate(args, 1):
        print(f'  {i}. {elem}')


resumir('Python', 'GeoPandas', 'Rasterio', 'OSMnx', titulo='Temario GeoPython 2026')
````

````python
# Variables locales vs globales
x = 'global'

def modificar_local():
    x = 'local'          # no afecta a la global
    print(f'Dentro: {x}')

modificar_local()
print(f'Fuera: {x}')     # sigue siendo 'global'
````

````python
# Lambdas: funciones anónimas de una línea
palabras = ['banana', 'kiwi', 'manzana', 'uva', 'cereza']

print(sorted(palabras, key=lambda p: len(p)))                     # por longitud
print(list(filter(lambda p: len(p) > 5, palabras)))               # más de 5 letras
print(list(map(lambda p: p.upper(), palabras)))                   # todas en mayúsculas
````

## 8. Manejo de errores

`try/except` evita que el programa se detenga al encontrar un error.

````python
datos = [10, 0, 5, 'texto', 8, None, 2]
resultados = []

for d in datos:
    try:
        resultados.append(100 / d)
    except ZeroDivisionError:
        print(f'División por cero')
    except TypeError as e:
        print(f'Tipo incorrecto ({d!r}): {e}')

print(resultados)
````

````python
# try / except / else / finally
def dividir(a, b):
    try:
        resultado = a / b
    except ZeroDivisionError:
        print('Error: división por cero')
        return None
    else:
        return resultado       # solo si NO hubo excepción
    finally:
        print('Esto se ejecuta siempre')


print(dividir(10, 2))
print()
print(dividir(10, 0))
````

````python
# Lanzar nuestras propias excepciones
def calcular_media(lista):
    if not lista:
        raise ValueError('La lista no puede estar vacía')
    if not all(isinstance(x, (int, float)) for x in lista):
        raise TypeError('Todos los elementos deben ser numéricos')
    return sum(lista) / len(lista)


for caso in [[1, 2, 3], [], [1, 2, 'tres']]:
    try:
        print(f'Media de {caso}: {calcular_media(caso)}')
    except (ValueError, TypeError) as e:
        print(f'Error: {e}')
````

## 9. Módulos y archivos

````python
%%writefile utilidades.py
"""
Módulo de utilidades — GeoPython 2026.
"""

def normalizar(valores):
    """Normaliza una lista al rango [0, 1]."""
    vmin, vmax = min(valores), max(valores)
    if vmin == vmax:
        return [0.0] * len(valores)
    return [(v - vmin) / (vmax - vmin) for v in valores]


def estadisticas(valores):
    """Devuelve media, min, max y desviación estándar."""
    import math
    n = len(valores)
    media = sum(valores) / n
    varianza = sum((v - media)**2 for v in valores) / n
    return {'n': n, 'media': round(media, 4),
            'min': min(valores), 'max': max(valores),
            'std': round(math.sqrt(varianza), 4)}
````

````python
import utilidades as ut

datos = [3, 7, 2, 9, 1, 5, 8, 4, 6]
print(ut.normalizar(datos))
print(ut.estadisticas(datos))
````

## 10. Ejercicio final: Rock, Paper, Scissors, Lizard, Spock

El juego popularizado por The Big Bang Theory. Lo implementamos aplicando todo lo visto hoy.

**Reglas:**
- Tijeras cortan Papel · Papel envuelve Piedra · Piedra aplasta Lagarto
- Lagarto envenena Spock · Spock destruye Tijeras · Tijeras decapitan Lagarto
- Lagarto come Papel · Papel refuta Spock · Spock vaporiza Piedra · Piedra aplasta Tijeras

````python
import random

OPCIONES = ['piedra', 'papel', 'tijeras', 'lagarto', 'spock']

# Diccionario (ganador, perdedor) → descripción
# Más limpio y escalable que un bloque de if/elif
REGLAS = {
    ('tijeras', 'papel'):   'Las tijeras cortan el papel',
    ('papel',   'piedra'):  'El papel envuelve la piedra',
    ('piedra',  'lagarto'): 'La piedra aplasta al lagarto',
    ('lagarto', 'spock'):   'El lagarto envenena a Spock',
    ('spock',   'tijeras'): 'Spock destruye las tijeras',
    ('tijeras', 'lagarto'): 'Las tijeras decapitan al lagarto',
    ('lagarto', 'papel'):   'El lagarto come el papel',
    ('papel',   'spock'):   'El papel refuta a Spock',
    ('spock',   'piedra'):  'Spock vaporiza la piedra',
    ('piedra',  'tijeras'): 'La piedra aplasta las tijeras',
}


def jugar_ronda(eleccion_jugador: str) -> dict:
    """
    Ejecuta una ronda contra la CPU.

    Returns:
        dict con jugador, cpu, resultado y descripcion
    """
    eleccion_jugador = eleccion_jugador.lower().strip()
    if eleccion_jugador not in OPCIONES:
        raise ValueError(f'Opción no válida. Elige entre: {OPCIONES}')

    eleccion_cpu = random.choice(OPCIONES)

    if eleccion_jugador == eleccion_cpu:
        resultado, descripcion = 'empate', '¡Empate!'
    elif (eleccion_jugador, eleccion_cpu) in REGLAS:
        resultado   = 'jugador'
        descripcion = REGLAS[(eleccion_jugador, eleccion_cpu)] + ' → ¡Ganas!'
    else:
        resultado   = 'cpu'
        descripcion = REGLAS[(eleccion_cpu, eleccion_jugador)] + ' → Pierdes.'

    return {'jugador': eleccion_jugador, 'cpu': eleccion_cpu,
            'resultado': resultado, 'descripcion': descripcion}
````

````python
class Partida:
    """Gestiona una partida con marcador, historial y estadísticas."""

    def __init__(self, nombre: str = 'Jugador'):
        self.nombre     = nombre
        self.victorias  = 0
        self.derrotas   = 0
        self.empates    = 0
        self._historial = []

    def __str__(self):
        return (f'Partida de {self.nombre} | '
                f'V:{self.victorias} D:{self.derrotas} E:{self.empates}')

    @property
    def total_rondas(self) -> int:
        return self.victorias + self.derrotas + self.empates

    @property
    def win_rate(self) -> float:
        return self.victorias / self.total_rondas * 100 if self.total_rondas else 0.0

    def jugar(self, eleccion: str) -> None:
        ronda = jugar_ronda(eleccion)
        self._historial.append(ronda)
        match ronda['resultado']:
            case 'jugador': self.victorias += 1
            case 'cpu':     self.derrotas  += 1
            case 'empate':  self.empates   += 1
        print(f"  Tú: {ronda['jugador']:8s} | CPU: {ronda['cpu']:8s} | {ronda['descripcion']}")

    def marcador(self) -> None:
        sep = '=' * 35
        print(f'\n{sep}\n  {self.nombre}\n{sep}')
        print(f'  Victorias: {self.victorias}')
        print(f'  Derrotas : {self.derrotas}')
        print(f'  Empates  : {self.empates}')
        print(f'  Win rate : {self.win_rate:.1f}%')
        print(sep)

    def eleccion_favorita(self) -> str:
        if not self._historial: return 'Sin datos'
        conteo = {op: 0 for op in OPCIONES}
        for r in self._historial: conteo[r['jugador']] += 1
        return max(conteo, key=conteo.get)

    def racha_maxima(self) -> int:
        racha = maximo = 0
        for r in self._historial:
            racha = racha + 1 if r['resultado'] == 'jugador' else 0
            maximo = max(maximo, racha)
        return maximo
````

````python
# Subclase con modo interactivo — herencia en acción
class PartidaInteractiva(Partida):
    """Extiende Partida para jugar pidiendo input al usuario."""

    def jugar_interactivo(self, n_rondas: int = 5) -> None:
        print(f'Opciones: {OPCIONES}\n')
        for i in range(1, n_rondas + 1):
            print(f'--- Ronda {i} ---')
            while True:
                eleccion = input('Tu elección: ').lower().strip()
                if eleccion in OPCIONES: break
                print(f'Opción no válida. Elige entre: {OPCIONES}')
            self.jugar(eleccion)
        self.marcador()
        print(f'  Favorita: {self.eleccion_favorita()}')
        print(f'  Racha máxima: {self.racha_maxima()}')
````

````python
# Demo automática
random.seed(2026)
partida = Partida('Sheldon')

print('=== PARTIDA ===')
for jugada in ['spock', 'spock', 'piedra', 'lagarto', 'papel',
               'tijeras', 'spock', 'lagarto', 'piedra', 'papel']:
    partida.jugar(jugada)

partida.marcador()
print(f'  Elección favorita: {partida.eleccion_favorita()}')
print(f'  Racha máxima:      {partida.racha_maxima()}')
````
